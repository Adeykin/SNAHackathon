```python
import pandas as pd
import numpy as np
import sklearn.metrics
```

```python
inputSubmission = '/home/adeykin/projects/SNAHackaton/out/imagesSubmit.csv'
outputSubmission = '/home/adeykin/projects/SNAHackaton/out/imagesSubmit_handReverse.csv'
```

```python
users = {}
f = open(inputSubmission, 'r')
for line in f.readlines():
    userIdStr, postsListStr = line.split(',', 1)
    userId = int(userIdStr)
    users[userId] = list(map(int, postsListStr[2:-3].split(',')))
```

```python
users
```

```python
#list(map(lambda a,b: reversed(b), users))
reversedUsers = dict(map(lambda x: (x[0], list(reversed(x[1]))), users.items()))
```

```python
f = open(outputSubmission, 'w')
for user, posts in reversedUsers.items():
    postsStr = map(str, posts)
    f.write( str(user) + ',"[' + ', '.join(postsStr) + ']"\n' )
```
