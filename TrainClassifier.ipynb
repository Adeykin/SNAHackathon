```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_pickle('trainWithtargetDF.pkl')
```

```python
df.head()
```

```python
X = df[df.columns[1:-1]].values
```

```python
y = df[df.columns[-1]].values
```

```python
del df
```

```python
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```
