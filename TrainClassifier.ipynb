```python
import pandas as pd
import numpy as np
import pickle
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
```

```python
#df = pd.read_pickle('trainWithtargetDF_fixed.pkl')
df = pd.read_pickle('trainWithtargetDF_fixed_normalised.pkl')
```

```python
df.head()
```

```python
X = df[df.columns[1:-1]].values
```

```python
y = df[df.columns[-1]].values
```

```python
del df
```

```python
X[X>0.6] = 1
X[X<=0.6] = 0
```

```python
y1 = y
y1[y1>0.5] = 1
y1[y1<=0.5] = 0
```

```python
y1.max()
```

```python
reg = GaussianNB().fit(X,y1)
reg.score(X, y1)
```

```python
pickle.dump(reg, open('gaussianBayes.pkl','wb'))
```

```python
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
pickle.dump(reg, open('linearClassifier.pkl','wb'))
```

# Validate on Train dataset

```python
yActual = reg.predict(X)
```

```python
newDF = df[['ImageId']]
```

```python
newDF['actualScore'] = yActual
```

```python
newDF.to_pickle('trainValidateOnTrainDF.pkl')
```
